```python
import pandas as pd
import requests
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv("Proyecto.csv") 
```

# Análisis de datos

#### Comprobamos los NaN y los sustituimos columna por columna por el mínimo

```python
nan_counts = df.isna().sum()
nan_counts
```

```python
df["Min_Salary"].fillna(df["Min_Salary"].mean(), inplace=True)
round(df,2)
```

```python
df["Max_Salary"].fillna(df["Max_Salary"].mean(), inplace=True)
round(df,2)
```

```python
median_salaries = round(df.groupby("Location")["Median_Salary"].mean().reset_index(),2)
median_salaries 
```

#### Importamos "import matplotlib.pyplot as plt" e "import seaborn as sns". Hago un gráfico de las ciudades y el salario medio que he sacado


```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Location", y="Median_Salary", data=median_salaries, palette="viridis")
plt.xticks(rotation=45)
plt.title("Salario Mediano por Ciudad")
plt.ylabel("Salario Mediano (USD)")
plt.xlabel("Ubicación")
plt.tight_layout()  
plt.show()
```

####  Estudio las empresas o portales que publican las ofertas para saber cuántas ofertas ofrecen cada uno y así saber quién tiene más ofertas publicadas en los datos de los que dispongo.

```python
publisher_counts = df["Publisher_Name"].value_counts()
publisher_counts 
```

```python
top_publisher = publisher_counts.idxmax()
top_publisher
```

```python
top_count = publisher_counts.max()
top_count 
```

#### Expongo las ofertas publicadas por Indeed que están en mis datos

```python
top_publisher_info = df[df["Publisher_Name"] == top_publisher]
top_publisher_info 
```

#### Analizo los datos que han salido en esta tabla sobre las ofertas de Indeed

```python
locations = top_publisher_info["Location"].unique()  
locations
```

```python
min_salary = top_publisher_info["Min_Salary"].tolist()
min_salary 
```

```python
max_salary = top_publisher_info["Max_Salary"].tolist() 
max_salary
```

```python
median_salary = top_publisher_info["Median_Salary"].tolist() 
median_salary 
```

#### Juntamos todo en la misma tabla y vemos ubicación y salarios.

```python
top_publisher_info_Indeed = df.groupby(["Location", "Median_Salary"])[["Min_Salary", "Max_Salary"]].mean().reset_index()
top_publisher_info_Indeed = round(top_publisher_info_Indeed, 2)
top_publisher_info_Indeed
```

 #### Estudio el orden de ciudades en relación con los tipos de salarios y anunciantes.

```python
grouped = round(df.groupby(["Location", "Publisher_Name"])["Min_Salary"].mean().reset_index(),2)
grouped
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped, x="Location", y="Min_Salary", hue="Publisher_Name", palette="viridis")

plt.title("Salario Mínimo Promedio por Ubicación y Editor", fontsize=16)
plt.xlabel("Ubicación", fontsize=14)
plt.ylabel("Salario Mínimo Promedio (USD)", fontsize=14)
plt.xticks(rotation=45)
plt.legend(title="Portal", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()
plt.show()
```

```python
grouped2 = round(df.groupby(["Location", "Publisher_Name"])["Median_Salary"].mean().reset_index(),2)
grouped2
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped2, x="Location", y="Median_Salary", hue="Publisher_Name", palette="viridis")

plt.title("Salario Medio Promedio por Ubicación y Editor", fontsize=16)
plt.xlabel("Ubicación", fontsize=14)
plt.ylabel("Salario Medio Promedio (USD)", fontsize=14)
plt.xticks(rotation=45)
plt.legend(title="Portal", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()
plt.show()
```

```python
grouped3 = round(df.groupby(["Location", "Publisher_Name"])["Max_Salary"].mean().reset_index(),2)
grouped3
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped3, x="Location", y="Max_Salary", hue="Publisher_Name", palette="viridis")

plt.title("Salario Máximo Promedio por Ubicación y Editor", fontsize=16)
plt.xlabel("Ubicación", fontsize=14)
plt.ylabel("Salario Máximo Promedio (USD)", fontsize=14)
plt.xticks(rotation=45)
plt.legend(title="Portal", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout()
plt.show()
```

# Parte final y conclusión

#### Agrupamos para comparar salario mediano y máximo con las localizaciones

```python
top_publisher_grouped = round(df.groupby(["Location", "Median_Salary"])[[ "Max_Salary"]].mean().reset_index(),2)
top_publisher_grouped
```

#### Ordeno por Median_Salary

```python
top_publisher_grouped_Max = top_publisher_grouped.sort_values(by="Median_Salary", ascending=False)
top_publisher_grouped_Max 
```

#### Ordeno por Max_Salary

```python
top_publisher_grouped_Max = top_publisher_grouped.sort_values(by="Max_Salary", ascending=False)
top_publisher_grouped_Max 
```

## Vemos que por salario máximo es más recomendable trabajar en Los Ángeles, pero que por salario mediano es más recomendable San Diego.
